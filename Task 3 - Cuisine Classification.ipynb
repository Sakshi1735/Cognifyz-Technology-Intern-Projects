```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

```

```python
data = pd.read_csv('Dataset .csv')
```

## Data Preprocessing

```python
data.head()
```

```python
data.isnull().sum()
```

```python
np.array(data['Cuisines'])
data.iloc[:,9:10]
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy='most_frequent')

imputer.fit(data.iloc[:,9:10].values)
data.iloc[:,9:10] = imputer.transform(data.iloc[:,9:10].values)
data.isnull().sum()
```

```python
#dropping unnecessary columns
columns_to_drop = [
    'Restaurant ID', 'Country Code', 'City', 'Address', 'Locality', 
    'Locality Verbose', 'Longitude', 'Latitude', 'Currency', 
    'Has Table booking', 'Has Online delivery', 'Is delivering now', 
    'Switch to order menu', 'Price range', 'Aggregate rating', 
    'Rating color', 'Rating text', 'Votes'
]

data.drop(columns=columns_to_drop, axis=1, inplace=True)
```

```python
data
```

```python
#Converting Categorical Data to Numerical
label_encoder = LabelEncoder() 
data['Restaurant Name'] = label_encoder.fit_transform(data['Restaurant Name']) 
data['Cuisines'] = label_encoder.fit_transform(data['Cuisines'])
```

```python
data
```

## Building Random Forest Model


```python
label_encoder_restaurant = LabelEncoder()
label_encoder_cuisines = LabelEncoder()
data['Restaurant Name'] = label_encoder_restaurant.fit_transform(data['Restaurant Name'])
data['Cuisines'] = label_encoder_cuisines.fit_transform(data['Cuisines'])
```

```python
X = data[['Restaurant Name', 'Average Cost for two']]
Y = data['Cuisines']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=10)
```

```python
model_DT = DecisionTreeClassifier(random_state=10, criterion="gini")
model_DT.fit(X_train, Y_train)
Y_pred = model_DT.predict(X_test)
```

```python
print("Actual vs Predicted:")
print(list(zip(Y_test, Y_pred)))

```

```python
# Evaluating the model
print("\nClassification Report:")
print(classification_report(Y_test, Y_pred))
```

```python
print("Accuracy Score:", accuracy_score(Y_test, Y_pred))
```

## Building Logistic Regression

```python
classifier = LogisticRegression(multi_class="multinomial")
#train the model object
classifier.fit(X_train,Y_train)      # fit is the function that is used for training the data
```

```python
Y_pred = classifier.predict(X_test)
print(Y_pred)
```

```python
cf_m = confusion_matrix(Y_test, Y_pred)
accuracy = accuracy_score(Y_test, Y_pred)
class_report = classification_report(Y_test, Y_pred)
```

```python
print("\nConfusion Matrix:")
print(cf_m)
```

```python
print("\nClassification Report:")
print(class_report)
```

```python
print("\nAccuracy of the model:")
print(accuracy)
```

## Conclusion

When comparing the two models, Random Forest demonstrates superior performance with an accuracy of 0.24071166928309787, compared to Logistic Regression's accuracy of 0.09419152276295134. This indicates that Random Forest is more effective at predicting the cuisines in this dataset, making it the better choice for this classification task.
